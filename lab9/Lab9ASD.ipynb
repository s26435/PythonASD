### Funkcja monożąca macierze:

```python
def mnozenieMacierzy(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Nie można pomnożyć tych macierzy. Liczba kolumn pierwszej macierzy nie jest równa liczbie wierszy drugiej macierzy.")
        return None

    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result
```

### Funkcja zwracająca macierz zerową:

```python
def macierz(m, n):
    return [[0 for _ in range(n)] for _ in range(m)]
```

### Funkcja wyświetlająca macierz:

```python
def wyswietlMacierz(A):
    for row in A:
        print(row)
```

Wykożystanie mnożenia macierzy:

```python
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]
wyswietlMacierz(mnozenieMacierzy(A,B))
```

Tworzenie macierzy zerowej:

```python
wyswietlMacierz(macierz(m=3,n=3))
```

### Mnożenie macierzy przez wektor:

```python
def wyswietlWektor(v):
    print("[", end="")
    for i in range(len(v)):
        print(v[i], end="")
        if i != len(v) - 1:
            print(", ", end="")
    print("]")

def mnozenieMacierzyPrzezWektor(matrix, vector):
    if len(matrix[0]) != len(vector):
        print("Nie można pomnożyć macierzy przez ten wektor. Liczba kolumn macierzy nie jest równa długości wektora.")
        return None

    result = [0 for _ in range(len(matrix))]

    for i in range(len(matrix)):
        for j in range(len(vector)):
            result[i] += matrix[i][j] * vector[j]

    return result
```

Przykład mnożenia macierzy przez wektor:

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

vector = [1, 2, 3]

wyswietlWektor(mnozenieMacierzyPrzezWektor(matrix, vector))
```
